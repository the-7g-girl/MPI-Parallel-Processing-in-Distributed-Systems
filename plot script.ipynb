```python
import matplotlib.pyplot as plt

# Data from the table
processes = ['8', '16', '32', '64']
read_times = [0.006876, 0.006508, 0.007912, 0.012689]
compute_times = [0.009785, 0.005515, 0.005149, 0.003238]
total_times = [0.017998, 0.013433, 0.014650, 0.017255]

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(processes, read_times, marker='o', label='Read Time')
plt.plot(processes, compute_times, marker='s', label='Compute Time')
plt.plot(processes, total_times, marker='^', label='Total Time')

# Labels and title
plt.xlabel('Number of Processes')
plt.ylabel('Time (seconds)')
plt.title('Average Execution Times vs. Number of Processes')
plt.legend()
plt.grid(True)
plt.xticks(processes)
# plt.xscale('log')

# Show the plot
plt.tight_layout()
plt.show()
```
